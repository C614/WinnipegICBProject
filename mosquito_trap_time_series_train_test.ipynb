```python
# This part changes 'no data' entries to NaN entries
for cols in list(df.columns):
    for indx in df.index:
        if df.at[indx, cols]=='no data':
            df.at[indx, cols]=np.NaN
```

```python
df.head()
```

```python
# changing everything else to date type
for cols in list(df.columns):
    if cols!='count_date':
        df[cols]=pd.to_numeric(df[cols])
```

```python
#filling the missing values using back-filling method for time series
df.north_west_1=df.north_west_1.fillna(method='bfill')
```

```python
#filling NW values 
df.north_west_2=df.north_west_2.fillna(method='bfill')
df.north_west_3=df.north_west_3.fillna(method='bfill')
df.north_west_4=df.north_west_4.fillna(method='bfill')
df.north_west_5=df.north_west_5.fillna(method='bfill')
df.north_west_6=df.north_west_6.fillna(method='bfill')
df.north_west_7=df.north_west_7.fillna(method='bfill')
```

```python
#filling SE values
df.south_east_1=df.south_east_1.fillna(method='bfill')
df.south_east_2=df.south_east_2.fillna(method='bfill')
df.south_east_3=df.south_east_3.fillna(method='bfill')
df.south_east_4=df.south_east_4.fillna(method='bfill')
df.south_east_5=df.south_east_5.fillna(method='bfill')
df.south_east_6=df.south_east_6.fillna(method='bfill')
df.south_east_7=df.south_east_7.fillna(method='bfill')
```

```python
#filling RR values
df.rural_aa=df.rural_aa.fillna(method='bfill')
df.rural_bb=df.rural_bb.fillna(method='bfill')
df.rural_cc=df.rural_cc.fillna(method='bfill')
df.rural_dd=df.rural_dd.fillna(method='bfill')
df.rural_ee=df.rural_ee.fillna(method='bfill')
df.rural_ff=df.rural_ff.fillna(method='bfill')
df.rural_gg=df.rural_gg.fillna(method='bfill')
df.rural_hh=df.rural_hh.fillna(method='bfill')
df.rural_ii=df.rural_ii.fillna(method='bfill')
```
